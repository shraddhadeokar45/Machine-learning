```python
import pandas as pd
import matplotlib.pyplot as plt

# Load iris dataset from URL
df = pd.read_csv("Iris.csv")
print(df.head())
```

```python
plt.figure(figsize=(15, 5))

# 1. Bar Graph: Mean SepalLengthCm per Species
plt.subplot(1, 3, 1)
mean_values = df.groupby('Species')['SepalLengthCm'].mean()
plt.bar(mean_values.index, mean_values.values, color=['orange', 'green', 'blue'])
plt.title('Mean Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')

# 2. Histogram: Petal Length
plt.subplot(1, 3, 2)
plt.hist(df['PetalLengthCm'], bins=20, color='purple', edgecolor='black')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')

# 3. Pie Chart: Species Distribution
plt.subplot(1, 3, 3)
species_count = df['Species'].value_counts()
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', colors=['orange', 'green', 'blue'])
plt.title('Species Distribution')

plt.tight_layout()
plt.show()
```
